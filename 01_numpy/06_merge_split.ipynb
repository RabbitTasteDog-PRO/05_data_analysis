# 병합/분리
- 배열을 합치거나 나누는 기능 정리임
- shape 조건 확인이 중요함


```python
import numpy as np
```

## 요소 삽입/삭제
- `np.insert()`: 지정 위치에 값 삽입함
- `np.delete()`: 지정 위치의 값 삭제함
- 원본 배열을 직접 바꾸지 않고 새 배열 반환함


```python
# np.insert(ndarray, index, elems, axis)
# 지정된 엘리먼츠를 삽입한 새 배열이 반환됨
# 원본 보잠
arr = np.arange(1, 13).reshape(3, 4)

# axis = 0 단위로 삽입
# == 행삽입
# - 삽입할 생의 열 개수는 삽입 대상 ndarray의 열 개수와 일치해야한다
print("insert(arr, 1, [50,50,50,50], axis=0):")
print(np.insert(arr, 1, [50, 50, 50, 50], axis=0))

print("np.insert(arr, 1, [70], axis=0)")
# 원소가 하나일경우 브로드케스팅으로 인해 행 맞춤
print(np.insert(arr, 1, [70], axis=0))

# 배열 기호 없어소 0차원이라 브로드케스팅 적용되서 들어감7
print("np.insert(arr, 1, 90, axis=0)")
print(np.insert(arr, 1, 90, axis=0))
```

```python
arr
# axis = 1 , 열방향
print("insert(arr, 1, [50,50,50,50], axis=0):")
print(np.insert(arr, 1, [50, 50, 50], axis=1))

print("np.insert(arr, 1, [70], axis=0)")
# 원소가 하나일경우 브로드케스팅으로 인해 행 맞춤
print(np.insert(arr, 2, [70], axis=1))

# 배열 기호 없어소 0차원이라 브로드케스팅 적용되서 들어감7
print("np.insert(arr, 1, 90, axis=0)")
print(np.insert(arr, 4, 90, axis=1))

# 1차운으로 변경 후 지정된 인덱스에 삽입
print("1차원으로 변경 후 np.insert(arr, 0, 100, axis=None)")
print(np.insert(arr, 0, 100, axis=None))
```

```python
# delete(ndarray, index, axis)
# - 원본을 직접 삭제하지 않고 결과가 적용된 새로운 배열 반환

print("arr")
print(arr)
print("np.delete(arr,0,axis=0)")
print(np.delete(arr,0,axis=0))
print("np.delete(arr,0,axis=1)")
print(np.delete(arr,0,axis=1))

# np.delete(arr, 0, 100, axis=None)
print("1차원으로 변경 후 print(np.delete(arr,0,axis=None))")
print(print(np.delete(arr,0,axis=None)))
```

## 병합
- `np.append()`: 배열 뒤에 값 또는 배열 추가함
- `np.vstack()`: 세로 방향 병합함
- `np.hstack()`: 가로 방향 병합함
- `np.concatenate()`: 지정 axis 기준 병합함
- `np.column_stack()`: 배열을 열 기준으로 결합함


```python
# 병합 함수는 shape조건이 맞아야만 수행 가능
# append
arr1 = np.arange(1,7).reshape(2,3)
arr2 = np.arange(7,13).reshape(2,3)

print(arr1)
print(arr2)

```

### np.append()
- 배열 뒤에 값을 추가함
- `axis=None`이면 1차원으로 펼쳐 붙임
- `axis=0/1` 지정 시 shape 조건 필요함


```python
print("np.append(arr1, arr2, axis=0)")
print(np.append(arr1, arr2, axis=0))

print("np.append(arr1, arr2, axis=1)")
print(np.append(arr1, arr2, axis=1))

print("np.append(arr1, arr2, axis=None)")
print(np.append(arr1, arr2, axis=None))
```

### np.vstack()
- vertical stack의 약자임
- 세로 방향으로 쌓음
- `axis=0` 방향 병합과 유사함
- 열 개수가 같아야 함


```python
print("np.vstack((arr1,arr2))")
print(np.vstack((arr1,arr2)))
```

### np.hstack()
- horizontal stack의 약자임
- 가로 방향으로 붙임
- `axis=1` 방향 병합과 유사함
- 행 개수가 같아야 함


```python
print("np.hstack((arr1,arr2))")
print(np.hstack((arr1,arr2)))
```

### np.concatenate()
- 가장 일반적인 배열 연결 함수임
- `axis=0`: 아래로 붙
- `axis=1`: 옆으로 붙
- `axis=None`: 1차원으로 펼쳐 붙


```python
print("np.concatenate((arr1, arr2), axis=0)")
print(np.concatenate((arr1, arr2), axis=0))

print("np.concatenate((arr1, arr2), axis=1)")
print(np.concatenate((arr1, arr2), axis=1))

print("np.concatenate((arr1, arr2), axis=None)")
print(np.concatenate((arr1, arr2), axis=None))
```

### np.column_stack()
- 1차원 배열을 열로 세워 결합함
- 여러 feature를 표 형태로 묶을 때 유용함


```python
# 1차원 배열을 열로 묶어서 2차원 배열로 만들어줌
a = np.array([1,2,3])
b = np.array([4,5,6])

print("np.column_stack((a,b))")
print(np.column_stack((a,b)))
```

## 분할
- `np.vsplit()`: 세로 방향 분할함
- `np.hsplit()`: 가로 방향 분할함
- 나누는 개수가 행/열 개수와 맞아야 함


```python
arr = np.arange(1, 25).reshape(6, 4)
arr
```

### np.vsplit()
- vertical split의 약자임
- 행 기준으로 배열을 나눔
- 행 개수가 나누는 개수로 정확히 나뉘어야 함



```python
# 분할할 수는 행 방향 == 행의 약수 (6,3,2,1)
# 분할할 수는 열 방향 == 열의 약수 (4,2,1)
print("np.vsplit(arr, 6)")
print("",np.vsplit(arr, 6))

print("np.vsplit(arr, 3)")
print("",np.vsplit(arr, 3))

print("np.vsplit(arr, 2)")
print("",np.vsplit(arr, 2))

# print("np.vsplit(arr, 4)")
# print("",np.vsplit(arr, 4))
```

### np.hsplit()
- horizontal split의 약자임
- 열 기준으로 배열을 나눔
- 열 개수가 나누는 개수로 정확히 나뉘어야 함


```python
print("np.hsplit(arr, 4)")
print("",np.hsplit(arr, 4))

print("np.hsplit(arr, 2)")
print("",np.hsplit(arr, 2))
```
