# 병합/분리
- 배열을 합치거나 나누는 기능 정리임
- shape 조건 확인이 중요함


```python
import numpy as np
```

## 요소 삽입/삭제
- `np.insert()`: 지정 위치에 값 삽입함
- `np.delete()`: 지정 위치의 값 삭제함
- 원본 배열을 직접 바꾸지 않고 새 배열 반환함


## 병합
- `np.append()`: 배열 뒤에 값 또는 배열 추가함
- `np.vstack()`: 세로 방향 병합함
- `np.hstack()`: 가로 방향 병합함
- `np.concatenate()`: 지정 axis 기준 병합함
- `np.column_stack()`: 배열을 열 기준으로 결합함


### np.append()
- 배열 뒤에 값을 추가함
- `axis=None`이면 1차원으로 펼쳐 붙임
- `axis=0/1` 지정 시 shape 조건 필요함


### np.vstack()
- vertical stack의 약자임
- 세로 방향으로 쌓음
- `axis=0` 방향 병합과 유사함
- 열 개수가 같아야 함


### np.hstack()
- horizontal stack의 약자임
- 가로 방향으로 붙임
- `axis=1` 방향 병합과 유사함
- 행 개수가 같아야 함


### np.concatenate()
- 가장 일반적인 배열 연결 함수임
- `axis=0`: 아래로 붙
- `axis=1`: 옆으로 붙
- `axis=None`: 1차원으로 펼쳐 붙


### np.column_stack()
- 1차원 배열을 열로 세워 결합함
- 여러 feature를 표 형태로 묶을 때 유용함


## 분할
- `np.vsplit()`: 세로 방향 분할함
- `np.hsplit()`: 가로 방향 분할함
- 나누는 개수가 행/열 개수와 맞아야 함


### np.vsplit()
- vertical split의 약자임
- 행 기준으로 배열을 나눔
- 행 개수가 나누는 개수로 정확히 나뉘어야 함


### np.hsplit()
- horizontal split의 약자임
- 열 기준으로 배열을 나눔
- 열 개수가 나누는 개수로 정확히 나뉘어야 함
