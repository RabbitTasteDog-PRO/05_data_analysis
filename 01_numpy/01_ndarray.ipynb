# numpy
- 수치 연산에 최적화된 Python 라이브러리임
- 다차원 배열 연산 처리 가능함
- 선형대수 등 다양한 수학 연산 지원함
- Python list 대비 시간/공간 효율이 좋음


## 환경설정
- NumPy 설치 및 버전 확인용 섹션임


## ndarray
- N-dimensional array의 약자임
- 다차원 데이터를 저장하고 연산하는 NumPy 배열 타입임
- 같은 배열 안의 원소는 하나의 `dtype`을 가짐


### 실습
- 과일 판매량 데이터 사용함
- NumPy 배열로 분기별 판매량 계산함
- 구할 값: 과일별 총 판매량, 과일별 평균 판매량임

- **데이터**
  - 1분기 판매량: `[200, 340, 150, 400]`임
  - 2분기 판매량: `[220, 300, 180, 420]`임
  - 과일 순서: 사과, 배, 오렌지, 바나나임


### ndarray 속성
- `shape`: 배열의 모양임
- `ndim`: 배열의 차원 수임
- `size`: 전체 원소 개수임
- `dtype`: 배열 원소의 데이터 타입임

#### dtype
- 배열 요소의 데이터 타입임
- ndarray는 한 배열 안에 하나의 dtype을 가짐

| 범주 | 종류 | 요약 |
|:---|:---|:---|
| 정수형 | `int8`, `int16`, `int32`, `int64` | 부호 있는 정수 타입임 |
| 정수형(부호 없음) | `uint8`, `uint16`, `uint32`, `uint64` | 음수 없는 정수 타입임 |
| 부동소수형 | `float16`, `float32`, `float64` | 실수 표현 타입임 |
| 복소수형 | `complex64`, `complex128` | 실수부와 허수부를 가지는 타입임 |
| 불리언형 | `bool_` | `True` 또는 `False` 표현 타입임 |
| 문자열형 | `string_`, `bytes_` | 고정 길이 문자열 타입임 |
| 유니코드형 | `unicode_` | 다국어 문자 표현 타입임 |
| 객체형 | `object_` | Python 객체 참조 저장 타입임 |
| 날짜/시간형 | `datetime64`, `timedelta64` | 날짜/시간 저장 타입임 |


### 특정 형식의 ndarray 생성
- `np.zeros(shape)`: 0으로 채워진 ndarray 생성함
- `np.ones(shape)`: 1로 채워진 ndarray 생성함
- `np.full(shape, n)`: 특정 값으로 채워진 ndarray 생성함
- `np.zeros_like(ndarray)`: 기존 배열과 같은 shape의 0 배열 생성함
- `np.ones_like(ndarray)`: 기존 배열과 같은 shape의 1 배열 생성함
- `np.full_like(ndarray, n)`: 기존 배열과 같은 shape의 특정 값 배열 생성함


### 특정 범위의 ndarray 생성
- `np.arange(start, stop, step)`: `start`부터 `stop` 전까지 `step` 간격으로 생성함
- `np.linspace(start, stop, num)`: `start`부터 `stop`까지 `num`개 균등 간격으로 생성함
- `np.logspace(start, stop, num, base)`: 지수 간격 값을 생성함
