# Indexing | Slicing
- indexing: 특정 위치의 값 선택임
- slicing: 범위 단위의 값 선택임
- fancy indexing: 여러 위치를 한 번에 선택하는 방식임
- boolean indexing: 조건이 `True`인 값만 선택하는 방식임


## indexing
- 인덱스 번호로 원소 하나를 선택함
- 0번부터 시작함
- 음수 인덱스 사용 가능함


## slicing
- `start:stop:step` 형식 사용함
- `start`는 포함함
- `stop`은 포함하지 않음
- `step`은 이동 간격임


## fancy indexing
- 인덱스 목록으로 여러 원소를 선택함
- 필요한 위치만 골라낼 때 사용함
- 반복문 없이 여러 값을 한 번에 추출 가능함


### 3차원 배열
- 배열 안에 배열이 중첩된 구조임
- 3차원부터는 공간 그림보다 참조 순서로 이해하는 것이 좋음
- 접근 순서: 바깥 축 → 행 → 열 순서임


### boolean을 이용한 fancy indexing
- 조건식 결과가 `True`인 원소만 선택함
- 데이터 필터링의 기본 방식임
- Pandas 조건 필터링과 연결되는 개념임
