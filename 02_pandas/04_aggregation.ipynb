```python
# 직급별 인원수 조회
print(emp_df['job'].value_counts())
emp_df.groupby('job')['name'].count()
```

```python
# 부서별 연봉그룹별 인원수 조회
# - 3000~4000, 4000~5000, 5000이상
# - 연봉그룹컬럼을 생성후 그룹핑
def set_salary_group(salary):
    if 3000 <= salary < 4000:
        return '3000대'
    elif 4000 <= salary < 5000:
        return '4000대'
    else:
        return '5000대이상'


# apply(함수): df, ser의 각 요소에 함수를 적용
emp_df['salary_group'] = emp_df['salary'].apply(set_salary_group)
# print(emp_df) #salary_group 컬럼 추가 확인

print(emp_df.groupby(['department', 'salary_group'])['name'].count())
print('-' * 30)
print(emp_df[['department', 'salary_group']].value_counts())
```

```python
# agg() 집계 함수 모음
# - 그룹화된 데이터에서 원하는 집계 연산을 적용하여 요약 통계를 반환한다
emp_df.groupby('department')['salary'].agg(['mean', 'min', 'max', 'std', 'count'])
```

```python
emp_df.groupby('department').agg({
    'salary': 'mean',
    'hire_date': 'min'
})
```

```python
dept_salary_df = emp_df.groupby("department", as_index=False).agg(
    employee_count=("name", "count"),
    avg_salary=("salary", "mean"),
    min_salary=("salary", "min"),
    max_salary=("salary", "max")
)

dept_salary_df["avg_salary"] = dept_salary_df["avg_salary"].round(1)
display(dept_salary_df)
```

## pivot_table
- 기본 기능: 데이터를 행, 열, 값으로 재구성하여 테이블을 만든다.
- 사용 방식: 단순히 재구성하려는 index, columns, values를 지정한다.
- 중복 허용: 동일한 index와 columns 조합이 중복될 경우, aggfunc를 통해 해당 중복된 데이터를 집계할 수 있다.
- 집계 함수 지원: aggfunc 매개변수를 사용하여 sum, mean, count 등의 집계 함수를 적용할 수 있다. (기본값: mean)
- 결측치 채움: fill_value 매개변수를 사용하여 결측치를 채울 수 있다.

```python
data = {
    '날짜': [
        '2024-01-01', '2024-01-01', '2024-01-01', '2024-01-01',
        '2024-01-02', '2024-01-02', '2024-01-02', '2024-01-02',
        '2024-01-03', '2024-01-03'
    ],
    '상품': [
        '아이폰 14', '갤럭시 S23', '아이패드 프로', '아이폰 14',
        '아이폰 14', '갤럭시 S23', '아이패드 프로', '갤럭시 S23',
        '맥북 프로', '아이폰 14'
    ],
    '판매량': [15, 30, 5, 10, 20, 25, 10, 5, 8, 12],
    '가격': [1000, 900, 800, 1000, 1000, 900, 800, 900, 2000, 1000]
}
sale_df = pd.DataFrame(data)
sale_df
```

```python
# pivot_table
sale_pivot_df = pd.pivot_table(
    sale_df,
    index='날짜',  # 행 기준을 날짜로 설정
    columns='상품',  # 열 기준을 상품으로 설정
    values='판매량',  # 표 안에 채울 값
    fill_value=0,  # 값이 없는 조합은 0으로 채움
    aggfunc='sum',  # 같은 날짜/상품이 이려개면 판매량을 합산
    margins=True  # 행/열 합계 추가
)

sale_pivot_df
```

## 타이타닉 데이터셋

```python
titanic_df = pd.read_csv('./data/titanic.csv')

# 성별, 선실등급(pclass)별 생존률
pivot_td = pd.pivot_table(
    titanic_df,
    index='Pclass',
    columns='Sex',
    values='Survived',
    aggfunc='mean',
    margins=True

).round(2).reset_index(drop=True)

display(pivot_td)

pivot_td = pd.pivot_table(
    titanic_df,
    index='Pclass',
    columns='Sex',
    values='Survived',
    aggfunc='mean',
    margins=True
).round(2)

pivot_td = pivot_td.reset_index()
pivot_td.columns.name = None

display(pivot_td)
```
