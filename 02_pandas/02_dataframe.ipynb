```python
# dict -> df생성
data = {"one": [1.0, 2.0, 3.0], "two": [10, 20, 30], "three": 99}

# dict의 각 item이 하나의 Series로 처리됨
df = pd.DataFrame(data)
print(df['one'])  # one 열에 있는 Series를 선택
df
```

```python
# 여러 개의 dict -> dataframe
data: list[dict[str, int]] = [{
    "a": 10,
    "b": 20,
}, {
    "a": 30,
    "b": 40,
    "c": 50,
}]

# 하나의 dict -> 하나의 레코드, key는 컬럼으로 처리
df = pd.DataFrame(data)
df  # 0번행의 c열의 NaN = 결측치
```

## DataFrame 생성

## 조회
- 행열조회
    - df.iloc[인덱스]
    - df.loc[라벨]
- 열조회
    - df[컬럼명]

```python
dicts = [
    {'name': '홍길동', 'job': '강사', 'age': 33},
    {'name': '신사임당', 'age': 44, 'job': '영업사원'},
    {'name': '이순신', 'age': 38, 'job': '장군'},
    {'name': '논개', 'age': 25, 'job': '개발'},
    {'name': '강감찬', 'age': 47, 'job': '교수'},
]
df = pd.DataFrame(dicts)
df
```

```python
print(df.iloc[0])
print(type(df.iloc[0]))  # 시리즈, 데이터프레임에서 한행 또는 한 열은 시리크 타입
print(df.loc[0].index)
print(df.iloc[0].name)
```

```python
# 슬라이싱 지워
df2 = df.iloc[0:2]
df2
```

### iloc (인덱스 위치)

```python
# fancy indexing 지원
df.iloc[[0, 2]]
```

```python
# 단일값 조회
print(df.iloc[0, 1])

# 여러 행 조회
df.iloc[0:2, 1]
```

```python
df2 = df.iloc[:2, :2]
print(type(df2))
df2
```

### loc
인데스라벨을 통해 조회

```python
# index명 지정 == 라벨링
df.index = ['client1', 'client2', 'client3', 'client4', 'client5']
df
```

```python
df.loc['client4']  # 1행만 나오므로 시리즈 타입
```

```python
# 인덱싱 하여 키를 제거하고 직접넣었기 때문에 key error 발생
# df.loc[3]
```

```python
# loc를 이룡한 컬럼 단일값 조회
print(type(df.loc['client2', 'job']))
print(df.loc['client2', 'job'])

# loc를 이용한 슬라이싱
print(df.loc['client2':'client4'])

print()
# loc를 이용한 슬라이싱
print(df.loc['client2':'client4', 'job'])

print()
# loc를 이용한 슬라이싱
print(df.loc['client2':'client4', 'job':])
```

### 컬럼라벨

```python
# df['컬럼명'] : 해당 컬럼 시리즈 반환
print(type(df['name']))
df['name']
```

```python
# fancy 인덱싱 : 조건값을 넣어서 팬시인덱승 가능
df[['name', 'job']]
```

```python
# loc + 컬럼 라벨
# print(df['name'])
print()
print(df['name'] == '논개')
print()

print(df.loc[df['name'] == '이순신'])
```

```python
# boolean indexing 비교연산
print(df['age'] > 30)
print('#' * 30)
print(df[df['age'] > 30])
```
