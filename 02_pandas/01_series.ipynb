## Series
시리즈(Series)는 인덱스(라벨)가 있는 1차원 배열 형태의 데이터 구조로, 데이터와 해당 데이터에 대한 인덱스를 함께 저장한다.

**주요 특징**
1. 시리즈는 데이터 및 관려 지수로 구성되며 숫자 인덱스가 자동으로 생성된다.
2. 인덱싱 : 라벨 기반 접근은 `loc`, 위치 기반 접근은 `iloc`을 사용한다. 각 데이터 포인트는 인덱스(라벨)을 통해 접근할 수 있다.
3. 동종 데이터 타입 : 시리즈 내의 모든 데이터는 동일한 데이터 타입이어야 한다.
4. 결측치(비어있는 값) 지원 : 시리즈는 결측치(NaN)을 지원하며 데이터가 누락된 경우에도 유연하게 처리
5. 다양한 데이터 소스 : 리스트, 배열, 딕셔너리 등 다양한 형식의 데이터를 기반으로 시리즈를 생성할 수 있다.

```python
stocks_ser = pd.Series(['NVDA', 'MSFT', 'AMZN', 'APPL', 'TSLA'], name='미국주식')  # 미국주식 티커
print(stocks_ser)
print(type(stocks_ser))
print(stocks_ser.index)
```

```python
# index를 지정하지 않은 시리즈 생성
# randn : 정규분포 표준편차 같이 생긴거
np.random.seed(1)
ser = pd.Series(np.random.randn(5))
print(ser)
print(ser[0], ser[1], ser[2], ser[3], ser[4])
print(ser.index)
```

```python
# index를 지정한 시리즈
# 라벨기반 접근 loc = 로케이션
ser = pd.Series(np.random.randn(5), index=['A', 'B', 'C', 'D', 'E'])
print(ser)
# error
# print(ser[0], ser[1], ser[2], ser[3], ser[4])
# 지정된 키로만 불러올 수 있다
print(ser['A'], ser['B'], ser['C'], ser['D'], ser['E'])
print(ser.index)
```

```python
# 라벨기반 접근 loc = 로케이션
ser = pd.Series(np.random.randn(5), index=['A', 'B', 'C', 'D', 'E'])
print("라벨기반접근 : ", ser.loc['A'], ser.loc['B'], ser.loc['C'])
# 인덱스 로케이션
print("위치기반접근 : ", ser.iloc[0], ser.iloc[1], ser.iloc[2])
```

```python
dct = {'a': 30, 'b': 40, 'c': 50}
ser = pd.Series(dct)
# print(ser)
ser
```

```python
# 날짜 인덱스 생성
date_index = pd.date_range('2026-05-28', periods=100)
# print(date_index)

# 정규분포를 이용한 시리즈
ser = pd.Series(np.random.randn(100), index=date_index)
# ser = pd.Series(np.random.normal(0,1,100), index=date_index)
# print(ser)
ser.name = '정규분포값'
print(ser.loc['2026-06-22'])
ser
```

### Series 속성

```python
movies = ['토이스토리5','눈동자','슈퍼걸','군체']
movies_ser = pd.Series(movies)

movies_ser
```

```python
print("dtype : ", movies_ser.dtype)
print("shape : ", movies_ser.shape)
print("size : ", movies_ser.size)
# 빈값 있다면 true
print("empty : ", movies_ser.empty)
# 중복값이 없다면 true
print("is_unique : ", movies_ser.is_unique)
print("index : ", movies_ser.index)
print("name : ", movies_ser.name)
```

### Series 메서드

```python
num_ser = pd.Series([300,100,500,200,400])
num_ser

print("num_ser.sum()", num_ser.sum())
print("num_ser.mean()", num_ser.mean())
print("num_ser.max()", num_ser.max())
print("num_ser.min()", num_ser.min())
print("num_ser.std()", num_ser.std())
print("num_ser.var()", num_ser.var())
print("num_ser.median()", num_ser.median())
print("num_ser.mode()", num_ser.mode()) # 최빈값
print("num_ser.count()", num_ser.count())
print("num_ser.unique()", num_ser.unique())
```

```python
print("num_ser.describe()", num_ser.describe())
print("num_ser.info", num_ser.info())
```

## S&P500
- Pandas는 다양한 파일 형식의 데이터를 **읽고** **작성**할 수 있다.
- 자주 사용하는 파일 형식으로 **CSV**, **Excel**, **JSON**, **SQL** 등이 있다.

```python
snp500_df = pd.read_csv('./data/S_P500_Prices.csv')
print(type(snp500_df))
print(snp500_df.shape)
print(snp500_df.duplicated().sum())

print("조회된 데이터 중 최상위 5개 : ", snp500_df.head())
print("조회된 데이터 중 최하위 5개 : ", snp500_df.tail())
# print(ser)
```

```python
# df -> ser 변환
# df.squeeze() : 컬럼이 1개인 df를 ser로 변환
snp500_ser = snp500_df.squeeze()
# print(snp500_ser.shape)
# snp500_ser.info()
snp500_ser

print("전체 요소 개수(size):", snp500_ser.size)
print("전체 요소 개수(len):", len(snp500_ser))
print("not-null 데이터 개수(count):", snp500_ser.count())

print("최솟값:", snp500_ser.min())
print("최댓값:", snp500_ser.max())
print("평균:", snp500_ser.mean())
print("중앙값:", snp500_ser.median())
print("분산:", snp500_ser.var())
print("표준편차:", snp500_ser.std())

# # 최빈치
# value, count = snp500_ser.mode()
# max_count = np.max(count)
# print(value[count == max_count])

print("통계 요약:")
snp500_ser.describe()
```
