# pandas

## Pandas 자료형
https://pandas.pydata.org/pandas-docs/stable/user_guide/dsintro.html

- DataFrame
  - 가로축과 세로축이 있는 엑셀과 유사한 데이터구조를 가진 자료형.
  - 가로축은 로우(행), 세로축은 컬럼(열)이다.
- Series
  - Pandas의 1차원 데이터 구조이다.
  - 데이터프레임의 각 컬럼은 모두 시리즈라는 자료형이다.
  - 데이터프레임은 n개의 시리즈로 구성된다.
  - 값(values)과 인덱스(index)를 함께 갖음.
  - 결측치(NaN)를 표현할 수 있음.

## Series
시리즈(Series)는 인덱스(라벨)가 있는 1차원 배열 형태의 데이터 구조로, 데이터와 해당 데이터에 대한 인덱스를 함께 저장한다.

**주요 특징**
1. 시리즈는 데이터 및 관려 지수로 구성되며 숫자 인덱스가 자동으로 생성된다.
2. 인덱싱 : 라벨 기반 접근은 `loc`, 위치 기반 접근은 `iloc`을 사용한다. 각 데이터 포인트는 인덱스(라벨)을 통해 접근할 수 있다.
3. 동종 데이터 타입 : 시리즈 내의 모든 데이터는 동일한 데이터 타입이어야 한다.
4. 결측치(비어있는 값) 지원 : 시리즈는 결측치(NaN)을 지원하며 데이터가 누락된 경우에도 유연하게 처리
5. 다양한 데이터 소스 : 리스트, 배열, 딕셔너리 등 다양한 형식의 데이터를 기반으로 시리즈를 생성할 수 있다.
